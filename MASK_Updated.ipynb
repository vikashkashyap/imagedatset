```python
# import the libraries as shown below

from tensorflow.keras.layers import Input, Lambda, Dense, Flatten
from tensorflow.keras.models import Model
from tensorflow.keras.applications.resnet50 import ResNet50
#from keras.applications.vgg16 import VGG16
from tensorflow.keras.applications.resnet50 import preprocess_input
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator,load_img
from tensorflow.keras.models import Sequential
import numpy as np
from glob import glob
import matplotlib.pyplot as plt
```

```python
# re-size all the images to this
IMAGE_SIZE = [224, 224]

train_path = '/content/imagedatset/Person data/train'
valid_path = '/content/imagedatset/Person data/test'
```

```python
# Import the Vgg 16 library as shown below and add preprocessing layer to the front of VGG
# Here we will be using imagenet weights

resnet = ResNet50(input_shape=IMAGE_SIZE + [3], weights='imagenet', include_top=False)
```

```python
# don't train existing weights
for layer in resnet.layers:
    layer.trainable = False
```

```python
# useful for getting number of output classes
folders = glob('/content/imagedatset/Person data/train/*')
```

```python
folders
```

```python
# our layers - you can add more if you want
x = Flatten()(resnet.output)
```

```python
x
```

```python
prediction = Dense(len(folders), activation='softmax')(x)

# create a model object
model = Model(inputs=resnet.input, outputs=prediction)
```

```python
# view the structure of the model
model.summary()
```

```python
?model.compile()
```

```python
# tell the model what cost and optimization method to use
model.compile(
  loss='categorical_crossentropy',
  optimizer='adam',
  metrics=['accuracy']
)
```

```python
# Use the Image Data Generator to import the images from the dataset
from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(rescale = 1./255,
                                   shear_range = 0.2,
                                   zoom_range = 0.2,
                                   horizontal_flip = True)

test_datagen = ImageDataGenerator(rescale = 1./255)
```

```python
# Make sure you provide the same target size as initialied for the image size
training_set = train_datagen.flow_from_directory('/content/imagedatset/Person data/train',
                                                 target_size = (224, 224),
                                                 batch_size = 32,
                                                 class_mode = 'categorical')
```

```python
training_set = training_set.map(lambda x_text, x_label: (x_text, tf.expand_dims(x_label, -1)))
test_set = test_set.map(lambda x_text, x_label: (x_text, tf.expand_dims(x_label, -1)))
```

```python
y_train = np.asarray(training_set.labels).astype('float32').reshape((-1,1))
y_test = np.asarray(test_set.labels).astype('float32').reshape((-1,1))
```

```python
model.fit()
```

```python
# fit the model
# Run the cell. It will take some time to execute
r = model.fit(
  x=training_set,
  validation_data=test_set,
  epochs=50,
  steps_per_epoch=len(training_set),
  validation_steps=len(test_set)
)
```

```python
# plot the loss
plt.plot(r.history['loss'], label='train loss')
plt.plot(r.history['val_loss'], label='val loss')
plt.legend()
plt.show()
plt.savefig('LossVal_loss')

# plot the accuracy
plt.plot(r.history['accuracy'], label='train acc')
plt.plot(r.history['val_accuracy'], label='val acc')
plt.legend()
plt.show()
plt.savefig('AccVal_acc')
```
